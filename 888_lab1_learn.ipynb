Line plot

```python
x=np.linspace(0,2*np.pi,50)
y1=np.sin(x)
y2=np.sin(2*x)
plt.figure()
plt.plot(y1)
plt.plot(x,y1)
```

```python
plt.plot(x,y1,'r')
```

```python
plt.plot(x,y1,'r-o')
```

```python
plt.clf()
plt.plot(x,y1,'g-o',x,y2,'b-+')
plt.legend(['sin(x)','sin(2x)'])
```

```python
plt.legend(['sin(x)','sin(2x)'])
```

Scatter plot

```python
N=50 #no of points
x=np.linspace(0,10,N)
#for printing x
from numpy.random import rand
e=rand(N)*5.0
y1=x+e
areas = rand(N)*300
plt.scatter(x,y1,s=areas)
```

```python
colors=rand(N) #calls colors as an attribute
plt.scatter(x,y1,s=areas,c=colors)
plt.colorbar() #just for colorbar
plt.title("random scatter")
```
