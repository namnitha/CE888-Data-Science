```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.linspace(0, 2*np.pi, 50)
x=np.sin(t)
y=np.cos(t)
```

```python
plt.figure()
plt.plot(x)
plt.plot(y)
```

```python
plt.figure()
plt.plot(y)
plt.title("cos")
```

Line plot

```python
x=np.linspace(0,2*np.pi,50)
y1=np.sin(x)
y2=np.sin(2*x)
plt.figure()
plt.plot(y1)
plt.plot(x,y1)
```

```python
plt.plot(x,y1,'r')
```

```python
plt.plot(x,y1,'r-o')
```

```python
plt.clf()
plt.plot(x,y1,'g-o',x,y2,'b-+')
plt.legend(['sin(x)','sin(2x)'])
```

```python
plt.legend(['sin(x)','sin(2x)'])
```

Scatter plot

```python
N=50 #no of points
x=np.linspace(0,10,N)
#for printing x
from numpy.random import rand
e=rand(N)*5.0
y1=x+e
areas = rand(N)*300
plt.scatter(x,y1,s=areas)
```

```python
colors=rand(N) #calls colors as an attribute
plt.scatter(x,y1,s=areas,c=colors)
plt.colorbar() #just for colorbar
plt.title("random scatter")
```

**Image plot- correlation plot**

```python
#creating the data
e1= rand(100)
e2= rand(100)*2
e3= rand(100)*50
e4= rand(100)*100
corrmatrix= np.corrcoef([e1,e2,e3,e4])
print(corrmatrix)
```

```python
#ploting corr matrix as image
plt.imshow(corrmatrix, cmap='GnBu')
plt.colorbar()
```

##Multiple plots using subplot

```python
t= np.linspace(0,2*np.pi)
x = np.sin(t)
y = np.cos(t)
plt.subplot(2,1,1)
plt.plot(x)
```

```python
#activating the new plot area
plt.subplot(2,1,2)
plt.plot(y)
```

##Histogram plot

```python
from numpy.random import randint
data = randint(10000, size=(10,1000))
x= np.sum(data, axis=0)   #approx norm distribution
plt.subplot(2,1,1)
plt.hist(x, color='r')
```

```python
#ploting cumulative dist 
plt.subplot(2,1,1)
plt.hist(x,cumulative=True)
```

##Legend, Title, Label axis

```python
#adding labels in the plot command
plt.plot(np.sin(t), label='sin')
plt.plot(np.cos(t), label='cos')
plt.legend() #legend is the corner small box which shows the names of the particular line
```

```python
plt.plot(t, np.sin(t))
plt.xlabel('radians')
plt.ylabel('amplitude', fontsize= 'medium')
plt.title('sin(x)', fontsize='large')
```

#NumPy Arrays Introduction

```python
import numpy as np
a = np.array([0,1,2,3,4,5])     #creating an array
print(a)
```

```python
print(type(a))    # checking the type
print(a.dtype)    #checking type of elements
```

```python
#checking dimensions and shape
print(a.ndim)
print(a.shape)
```

```python
print(a.itemsize)   #checking bytes per element
print(a.nbytes)     #for bytes of memory use
```

###Array operations

```python
a = np.array([1, 2, 3, 4])
b = np.array([2, 3, 4, 5])

print(a*b)
print(a**b)  #a power b
print(a+b)
print(a/b)
```

```python
print(a[0])
```

```python
#setting the array elements
a[0]=10
print(a[0])
print(a)
```

```python
# Multi-dimensional array

a = np.array([[ 0, 1, 2, 3],
              [10,11,12,13]])
print(a)
```

```python
print(a.shape)
print(a.size)
print(a.ndim)
```

```python
print(a[1,3])
```

```python
#get and set multi dimentional array
a[1,3]=-11
print(a[1,3])
```

```python
print(a[1])
```

##Slicing arrays

```python
print(a)
```

```python
print(a[:,2]) #prints all row column2
print(a[0,:]) #prints all columns of row0
print(a[:,:-2]) #prints all rows with all colums of second last
```
